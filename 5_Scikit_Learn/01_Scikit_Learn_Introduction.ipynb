### My Understanding

Scikit-Learn is a Python library that gives us ready-made machine learning algorithms.

Without it, we'd have to implement algorithms like Linear Regression ourselves.

With Scikit-Learn, we can import them and focus on training and prediction.

## Install vs Import

Install:
Downloads and adds the package to the Python environment.

Import:
Makes the installed package available in the current Python program.

## What is __version__?

__version__ is a special attribute that stores the installed version of the library.

It is an attribute, not a function.
