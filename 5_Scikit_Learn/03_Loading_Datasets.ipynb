```python
from sklearn.datasets import load_iris
```

### load_iris
It imports the load_iris function from sklearn.datasets so we can load the built-in Iris dataset later.
- A built-in function in Scikit-Learn.
- Used to load the Iris dataset.
- Returns the dataset into Python.

```python
iris = load_iris()
```

### Bunch Object

The Iris dataset is stored as a **Bunch** object.

A Bunch is a special Scikit-Learn container that stores multiple related pieces of information together, such as:

- data
- target
- feature names
- target names
- description

```python
print(type(iris))
```

```python
print(iris.keys())
```

### Important Keys

The most commonly used keys are:

- `data` → Input features (X)
- `target` → Output labels (y)
- `feature_names` → Names of the input features
- `target_names` → Names of the output classes

```python
print(iris.data)
```

### iris.data

- Each row represents one flower.
- Each column represents one feature (measurement).
- The Iris dataset contains four features for every flower.

```python
print(iris.data.shape)
```

### .shape

- `.shape` returns the dimensions of the dataset.
- The first value represents the number of rows (samples).
- The second value represents the number of columns (features).
- It returns the result as a tuple.

```python
print(iris.feature_names)
```

### feature_names

- `feature_names` returns the names of the input features (columns).
- Each feature describes one characteristic of a flower.
- In the Iris dataset, there are four input features:
  - Sepal Length
  - Sepal Width
  - Petal Length
  - Petal Width

```python
print(iris.target)
```

### target

- `target` stores the output labels (y).
- Each value corresponds to one row in `iris.data`.
- The Iris dataset uses numeric class labels:
  - 0
  - 1
  - 2
- Every sample has exactly one target label.

```python
print(iris.target_names)
```

### target_names

- Each numeric label in `iris.target` maps to a class name.
- Mapping:
  - 0 → setosa
  - 1 → versicolor
  - 2 → virginica

```python
print(iris.data[0])
```

### iris.data[0]

- `[0]` accesses the first row of the dataset.
- The first row represents the first flower.
- It returns all four feature values of that flower.

```python
print(iris.target[0])
```

```python
print(iris.target_names[iris.target[0]])
```

### Accessing a Target Name

- `iris.target[0]` returns the class label of the first sample.
- `iris.target_names[iris.target[0]]` converts the numeric label into the actual class name.
- This helps us understand what each numeric label represents.
