# Introduction to NumPy

### What is NumPy?

- **NumPy (Numerical Python)** is a Python library for fast numerical computing.
- It provides a powerful data structure called **ndarray (N-dimensional Array)**.
- NumPy is faster and more memory-efficient than Python lists for numerical data.
- It is the foundation of libraries like **Pandas, Matplotlib, Scikit-learn, TensorFlow, and PyTorch**.

### Importing NumPy

```python
import numpy as np
```

- `np` is the standard alias used by the Python community.
- It makes the code shorter and easier to read.

````python
import numpy as np
print(np.__version__)
````

## Why NumPy?

Python lists are flexible but slower for numerical computations.

**Python List**
- Can store multiple data types
- Slower
- More memory

**NumPy Array**
- Stores one data type
- Faster
- Memory efficient

````python
numbers = [10,"Hello",3.14,True]
print(numbers)
````

````python
print(type(numbers))
````

````python
arr = np.array([10,20,30,])
print(arr)
print(type(arr))
````

ndarray = N-dimensional Array

````python
A = np.array([1,2,3])
B = np.array([[1,2,3]])

print(A.ndim)
print(B.ndim)
````

````python
A = np.array(10)
print(A)
print(A.ndim)
````

## Array Dimensions

- 0D → Single value
- 1D → One-dimensional array (vector)
- 2D → Two-dimensional array (rows and columns)
- 3D → Collection of 2D arrays
- `ndim` returns the number of dimensions of an array.

````python
A = np.array([
    [10,20,30],
    [40,50,60]
])
print(A.shape)
````

````python
print(type(A.shape))
````
